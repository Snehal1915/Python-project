### 2. Answer the following questions

##### A. Create a new column called job_title with the below conditions (7 Marks)

**Step1:** Create a user define function for converting the below job title names into lower case (5 Marks)

    Data Scientist = data scientist

    Data Engineer  = data engineer

    Data Analyst = data analyst

    Machine Learning Engineer = mle

    Manager = manager

    Director = director

    The other titles = other

**Step2:** Display the count plot for job_title (2 Marks)



```python
 def convert(x):
    if x.strip() in ['Data Scientist','Data Engineer','Data Analyst','Manager','Director']:
        return x.lower()
    if x=='The other titles':
        return 'other'
    if x=='Machine Learning Engineer':
        return 'mle'
    else:
        return 'other'
```

```python
df['Job_title']=df['Job Title'].map(convert)
df['Job_title']
```

```python
sns.countplot(df['Job_title'])
```


#### B. Create a new column called job_seniority with the below conditions (6 Marks)


**Step1:** Create a user define function for converting the below job title names into lower case (4 Marks)

    sr = senior or lead or principal

    jr = junior or jr.

    The other titles = other

**Step2:** Display the count plot for job_seniority (2 Marks)


```python
def job(x):
    if 'senior' in x.lower() or 'lead' in x.lower() or 'principle' in x.lower():
        return 'sr'
    elif 'junior' in x.lower():
        return 'jr'
    else:
        return 'others'
```

```python
df['job_seniority']=df['Job Title'].map(job)
df['job_seniority'].unique()
```

```python
sns.countplot(df['job_seniority'])
```

### 3. Answer the following questions using `salary` column ()



* Step1:Create a new column called `salary` from `Salary Estimate` with the below (3 marks)



**Condition:** Extract from 53 91K (Glassdoor est.) into $53K-$91K



* Step2: Replace -1 value with NaN (2 Mark)



* Step3: Removing '$' and 'K' (2 Mark)



* Step4: Remove the string employer provided salary from the salary array (3 marks)



**Input:**'Employer Provided Salary:120-140'



**Output:** '120-140'

```python
#Step 1
df['Salary']=df['Salary Estimate'].map(lambda x:'53K-91K' if x=='$53K-$91K (Glassdoor est.)' else x)
df['Salary']
```

```python
# Step 2
df['Salary'].replace({'-1':np.NaN}, inplace=True)
```

```python
df['Salary'].isnull().sum()
```

```python
#Step 3
df['Salary']=df['Salary'].map(lambda x : str(x).replace('$',''))
df['Salary']=df['Salary'].map(lambda x : str(x).replace('K',''))
```

```python
df['Salary'].head()
```

```python
#Step 4
df['Salary']=df['Salary'].map(lambda x: str(x).replace('Employer Provided Salary:',''))
df['Salary']=df['Salary'].map(lambda x: str(x).replace('Per Hour',''))
df['Salary'].head()
```

```python
#Step 5
df1=df[df['Salary Estimate'].str.contains('Per Hour')]['Salary'].map(lambda x:str(float(x.split('-')[0])*8*24*12)+'-'+
                                                                    str(float(x.split('-')[0])*8*24*12))
```

```python
df1
```

```python
#Step 6
df['Salary']=df['Salary'].map(lambda x:x.split(' ')[0])
```

```python
df['Salary1']=df['Salary'].map(lambda x:(float(x.split('-')[0])))
df['Salary2']=df['Salary'].map(lambda x: x.replace('(Employer','')).replace('nan','0-0').map(lambda x:(float(x.split('-')[1])))
```

```python
df['Salary']=(df['Salary1']+df['Salary2'])/2
df['Salary'].head()
```

```python
#Step 7
df['Salary'].fillna(df['Salary'].mean(),inplace=True)
```
